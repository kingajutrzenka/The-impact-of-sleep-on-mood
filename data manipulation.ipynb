```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_data.csv', sep=';')
df.head()
```

*The impact of Total Sleep Hours on Mood Score, Stress Level and Productivity Score*

```python
correlation = df['Total Sleep Hours'].corr(df['Mood Score'])
sns.scatterplot(x='Total Sleep Hours', y='Mood Score', data=df)
print(f"Correlation between Total Sleep Hours and Mood Score: {correlation}")
print("It is very weak correlation")
```

```python
correlation = df['Total Sleep Hours'].corr(df['Stress Level'])
sns.scatterplot(x='Total Sleep Hours', y='Stress Level', data=df)
print(f"Correlation between Total Sleep Hours and Stress Level: {correlation}")
print("It is very weak correlation")
```

```python
correlation = df['Total Sleep Hours'].corr(df['Productivity Score'])
sns.scatterplot(x='Total Sleep Hours', y='Productivity Score', data=df)
print(f"Correlation between Total Sleep Hours and Productivity Score: {correlation}")
print("It is very weak correlation")
```

*The impact of Sleep Quality on Mood Score, Stress Level and Productivity Score*

```python
correlation = df['Sleep Quality'].corr(df['Mood Score'])
sns.scatterplot(x='Sleep Quality', y='Mood Score', data=df)
print(f"Correlation between Sleep Quality and Mood Score: {correlation}")
print("It is very weak correlation")
```

```python
correlation = df['Sleep Quality'].corr(df['Stress Level'])
sns.scatterplot(x='Sleep Quality', y='Stress Level', data=df)
print(f"Correlation between Sleep Quality and Stress Level: {correlation}")
print("It is very weak correlation")
```

```python
correlation = df['Sleep Quality'].corr(df['Productivity Score'])
sns.scatterplot(x='Sleep Quality', y='Productivity Score', data=df)
print(f"Correlation between Sleep Quality and Productivity Score: {correlation}")
print("It is very weak correlation")
```

*All correlation*

```python
correlation_matrix = df.select_dtypes(include=np.number).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

The heatmap shows no significant correlations between the selected variables. All correlation values are close to zero, indicating no linear relationships between sleep - related factors and psychological scores such as mood, stress, or productivity.
